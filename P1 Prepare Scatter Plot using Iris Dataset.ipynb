## Import Libraries

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

## Import Dataset

```python
df = pd.read_csv('IRIS.csv')
df.tail()
```

## Making Scatter Plot (for each feature vs species)

```python
features = list(df.columns)
print(features)

species = features.pop(4)
print(species)
```

```python
for feature in features:
    plt.scatter(species,feature, data = df)
    plt.title(feature+ ' Vs ' + 'Species')
    plt.xlabel('species')
    plt.ylabel(feature)
    plt.show()
```
